```python
numbers = [7,30,93,55]
x = input("Enter number (or done): ")
while x != "done":
    numbers.append(int(x))
    x = input("Enter number (or done): ")

print("List", numbers)
print("Ascending", sorted(numbers))
print("Descending", sorted(numbers, reverse=True))
```

```python
y = [100, 50, 330, 50, 660]

def modify_list(y):
    y.append(100)
    y.insert(2, 999)

    if 50 in y:
        y.remove(50)

    y.pop()
    return y

print(modify_list(y))
```

```python
students = []

for i in range(5):
    name = input("Enter student name: ")
    grade = int(input("Enter grade: "))
    students.append([name, grade])

highest = students[0]
lowest = students[0]

for student in students:
    if student[1] > highest[1]:
        highest = student
    if student[1] < lowest[1]:
        lowest = student

print("Highest grade:", highest[0], highest[1])
print("Lowest grade:", lowest[0], lowest[1])
```

```python
temperatures = [17, 33, 22, 30, 11, 17, 13, 8, 27, 32]

above_30 = 0
below_15 = 0

for temp in temperatures:
    if temp > 30:
        above_30 += 1
    if temp < 15:
        below_15 += 1

print("Number of temperatures above 30:", above_30)
print("Number of temperatures below 15:", below_15)
```

```python
def remove_all_occurrences(lst, value):
    while value in lst:
        lst.remove(value)
    return lst
```

```python
secret_number = 42
attempts = 0

while True:
    guess = int(input("Guess the number: "))
    attempts += 1

    if guess > secret_number:
        print("Too high")
    elif guess < secret_number:
        print("Too low")
    else:
        print("Correct!")
        print("Number of attempts:", attempts)
        break
```
