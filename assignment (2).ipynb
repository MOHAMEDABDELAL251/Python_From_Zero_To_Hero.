##1) Write a program that takes a list of numbers from the user (until they enter ‘done’), then prints the list, its sorted version (ascending and descending)



```python
numbers = [7,30,93,55]
x = input("Enter number (or done): ")
while x != "done":
    numbers.append(int(x))
    x = input("Enter number (or done): ")

print("List", numbers)
print("Ascending", sorted(numbers))
print("Descending", sorted(numbers, reverse=True))
```

#2) Create a function modify_list(lst) that receives a list, appends 100 to it, inserts 999 at index 2, removes the first occurrence of 50 (if exists), pops the last element, and returns the modified list.

```python
y = [100, 50, 330, 50, 660]

def modify_list(y):
    y.append(100)
    y.insert(2, 999)

    if 50 in y:
        y.remove(50)

    y.pop()
    return y

print(modify_list(y))
```

#3) Write a program that asks the user for 5 student names and their grades, stores them in a list of lists (e.g. [[‘Ali’, 85], [‘Sara’, 92]]), then prints the student with the highest and lowest grade.

```python
students = []

for i in range(5):
    name = input("Enter student name: ")
    grade = int(input("Enter grade: "))
    students.append([name, grade])

highest = students[0]
lowest = students[0]

for student in students:
    if student[1] > highest[1]:
        highest = student
    if student[1] < lowest[1]:
        lowest = student

print("Highest grade:", highest[0], highest[1])
print("Lowest grade:", lowest[0], lowest[1])
```

#4) Create a function modify_fruits(fruits) that takes a list of fruits, does the following in order:

Appends “kiwi”
-
Inserts “grape” at index 2
-
Changes the item at index 0 to
uppercase version
-
Removes the second item (by index) using pop()
-
Returns the modified list

```python
fruits=["apple","mango","orange","cherry","lemon"]

def my_function (fruits) :
  fruits.append("kiwi")
  fruits.insert(2,"grape")
  fruits[0]=fruits[0].upper()
  fruits.pop(2)
  return fruits
print(my_function(fruits))
```

#6) Implement a program that creates a list of temperatures (you choose 10 values), then uses a for loop to count how many are above 30 and how many below 15, and prints both counts.

```python
temperatures = [17, 33, 22, 30, 11, 17, 13, 8, 27, 32]

above_30 = 0
below_15 = 0

for temp in temperatures:
    if temp > 30:
        above_30 += 1
    if temp < 15:
        below_15 += 1

print("Number of temperatures above 30:", above_30)
print("Number of temperatures below 15:", below_15)
```

#7) Create a function remove_all_occurrences(lst, value) that removes all occurrences of a given value from the list

```python
def remove_all_occurrences(lst, value):
    while value in lst:
        lst.remove(value)
    return lst
```

#8) Write a number guessing game: computer has a fixed number (e.g. 42), user guesses via input() in a while loop, give hints “too high”/”too low”, count attempts, end when correct.

```python
secret_number = 42
attempts = 0

while True:
    guess = int(input("Guess the number: "))
    attempts += 1

    if guess > secret_number:
        print("Too high")
    elif guess < secret_number:
        print("Too low")
    else:
        print("Correct!")
        print("Number of attempts:", attempts)
        break
```

#9) Write a program that asks the user to enter 7 city names, stores them in a list, then uses a for loop to print each city with its length (e.g. “Cairo – 5 letters”).

```python
cities = []

for i in range(7):
    city = input("Enter city name: ")
    cities.append(city)

print("\nCities and their lengths:")
for city in cities:

    print(city, "-", len(city), "letters")
```
